# Problem 1
Vehicle sınıfından miras alan bir Bus sınıfı oluşturun. Bus.seating_capacity()'nin kapasite bağımsız değişkenine varsayılan olarak 50 değerini verin.

```
Output:
> The seating capacity of a bus is 50 passengers
```


````python
class vehicle():
    def __init__(self,marka,renk):
        self.marka = marka
        self.renk = renk

class bus(vehicle):
    def __init__(self,marka,renk,kapasite) :
        super().__init__(marka,renk)
        self.kapasite = kapasite
    
    def seating_capacity(self):
        print(f"The seating capasity of a bus is {self.kapasite} passengers")

bus1 = bus("Audi","Beyaz",50)
bus1.seating_capacity()

````
