# Problem 2 Faktöriyel

Kullanıcıdan bir sayı alınız ve alınan sayının faktöriyelini hesaplayan bir program yazınız. 


```python
def factoriyel(n):
    if n == 0:
        return 1
    else:
        return n*factoriyel(n-1)
sayi = int(input("Faktöriyelini hesaplamak istediğiniz sayiyi giriniz:"))
print(factoriyel(sayi))
```

# Problem 3 Virgül ile Ayırma

Kullanıcıdan bir metin halinde virgül ile ayrılmış sayılar alınız ve aldığınız her bir sayıyı listeye atayınız, daha sonra listedeki her elemanı toplayıp ekrana bastıran bir program yazınız.

<h4>Sample Input</h4>
<p>
Lutfen sayilari virgul ile ayrirarak giriniz: 5,14,7,2,1
</p>
<h4>Output</h4>

Toplam: 29

```python
metin=input("Virgül ile ayrilmiş sayilari giriniz:")
sayilar = metin.split(",") #metin stringini virgüle göre böler ve  sayıları bir liste olarak saklar. 2,4,5,6,7 metin halşnde aldık listeye çevirdik['2', '4', '5', '6', '7']
toplam=0
for i in sayilar:
    toplam+=int(i) # Her bir sayıyı integer tipine dönüştürüp toplama ekler
print("Toplam:",toplam)

```

# Problem 5 Ürün Kayıt Uygulaması
Programımız ilk olarak kullanıcıdan, kaydı yapılacak ürünlerin sayısını isteyecek. Ardından yazılmış olan ürün sayısı kadar ürün ismi ve ürün fiyatı bilgisini kullanıcıdan alacak. Son olarak ise ürünleri ekrana yazdıracağız.

<h4>Sample Input:</h4>
<p>
Lutfen kac adet urun gireceginizi yaziniz: 3
<br/>
1. Urununuzun ismini giriniz: Mentos
<br/>
1. Urununuzun fiyatini giriniz: 1
<br/>
2. Urununuzun ismini giriniz: Meybuz
<br/>
2. Urununuzun fiyatini giriniz: 0.25
<br/>
3. Urununuzun ismini giriniz: Uzun Şeker
<br/>
3. Urununuzun fiyatini giriniz: 0.75
<br/>
<h4>Output:</h4>
1. Urun: Mentos - 1 TL
<br/>
2. Urun: Meybuz - 0.25 TL
<br/>
3. Urun: Uzun Şeker - 0.75 TL
</p>

```python
urun_sayisi=int(input("Lütfen kaç adet ürün girecegiizi yaziniz:"))
urunler={}
for i in range(urun_sayisi):
    x=input( "{}.Urununuzun ismini giriniz:".format(i+1))
    y=float(input( "{}.Urununuzun fiyatini giriniz:".format(i+1)))
    urunler[x]=y
    

print("Kaydedilen ürünler:")
for x, y in urunler.items():
    print(f"{x}: {y} TL")
```

# Problem 6 Türkçe Karakter Çevirici
Programa girilen dosya ismindeki Türkçe karakterleri, İngilizce'ye çeviren programı yazınız.

<h4>Sample Input:</h4>
<p>
Lutfen dosya ismi giriniz: fotoğraf.jpeg
<br>
Dosya ismi: fotograf.jpeg
<br>
<br>
Lutfen dosya ismi giriniz: ışık.png
<br>
Dosya ismi: isik.png
</p>

```python
dosya_adi = input("Lütfen dosya adını girin: ")

# Türkçe karakterleri İngilizce karakterlere dönüştür
turkce_karakterler = "çğıöşüÇĞİÖŞÜ"
ingilizce_karakterler = "cgiosuCGIOSU"
ceviri_tablosu = str.maketrans(turkce_karakterler, ingilizce_karakterler)#str.maketrans() yöntemini kullanarak Türkçe karakterleri İngilizce karakterlere dönüştürmek için bir çeviri tablosu oluşturur
dosya_adi = dosya_adi.translate(ceviri_tablosu)#translate() yöntemi kullanarak girilen dosya adındaki Türkçe karakterleri bu çeviri tablosuyla İngilizce karakterlere dönüştürür.

print("İngilizce karakterlere dönüştürülmüş dosya adı: ", dosya_adi)
```

# Problem 7 Sayı Tahmin Oyunu ama Bilgisayar Tahmin Ediyor :)
Sayı tahmin oyunu: İki oyuncudan birinin, 0 ile 100 arasında bir sayı tutması ile başlar. Diğer oyuncu ise tahminler yaparak tutulan sayıyı bulmaya çalışır ve yaptığı her tahminde sayıyı tutmuş olan oyuncudan YUKARI veya AŞAĞI olarak geri bildirim alır. Sayıyı bildiğinde ise rakip oyuncu kaç tahminde (kaç denemede) bildiğini söyler.

Senden yazmanı beklediğimiz program ise Python'ın tahminleri yapan tarafta olacağı bir oyun oluşturman. Yani sen 0 ile 100 arasında bir sayı tutacaksın ardından program senin tuttuğunuz sayıyı bulmaya çalışacak.
<h4>Sample Input:</h4>
Bu örnekte aklınızdan 90 tuttuğunuzu varsayalım.
<p>
Oyun başlıyor lutfen aklindan bir sayi tut! Tuttun mu, hazır mısın? : 1
<br>
Tahmin: 50
<br>
Durum: Yukari
<br>
Tahmin: 75
<br>
Durum: Yukari
<br>
Tahmin: 87
<br>
Durum: Yukari
<br>
Tahmin: 93
<br>
Durum: Asagi
<br>
Tahmin: 90
<br>
Durum: Bildin
</p>

```python
import random
import time
print("""  
Oyun başlıyor lütfen aklından bir sayi tut. """)

tahmin_hakki=5
benim_tuttugum= 50

while True:
    
    tahmin=random.randint(1,100)
    if tahmin < benim_tuttugum:
        print("Tahmin:",tahmin)
        time.sleep(1)
        print("Durum: Yukari")
        tahmin += random.randint(1,100)
        tahmin_hakki -= 1
    elif tahmin > benim_tuttugum:
        print("Tahmin:", tahmin)
        time.sleep(1)
        print("Durum:Asagi")
        tahmin -= random.randint(1,100)
        tahmin_hakki -= 1
    else:
        print("Tahmin:", tahmin)
        time.sleep(1)
        print("Durum : Bildin")
        break
    if (tahmin_hakki==0):
        print("Tahmin hakkınız bitti")
        print("Sayimiz:", benim_tuttugum)
        break

```

# Problem 9 Tek-Çift

Kullanıcıdan 'q' tuşuna basana kadar rastgele sayılar alın. Tek olanları bir listeye, çift olanları bir başka listeye büyükten küçüğe sıralanacak şekilde atın. Listeleri ekrana yazdıran bir kod yazın.

```python
tek_sayilar = []
cift_sayilar = []

while True:
    sayi = input("Lütfen bir sayı girin ('q' çıkış yapmak için): ")
    
    if sayi == 'q':
        break
    
    sayi = int(sayi)
    
    if sayi % 2 == 0:
        cift_sayilar.append(sayi)
    else:
        tek_sayilar.append(sayi)
        
tek_sayilar.sort(reverse=True) #REVERSE = TRUE büyükten küçüğe sıralamamızı sağlar. REVERSE = FALSE küçükten büyüğe sıralmamız sağlar.İkisi sort fonksiyonun parametreidir.
cift_sayilar.sort(reverse=True)

print("Tek sayılar: ", tek_sayilar)
print("Çift sayılar: ", cift_sayilar)
```

 # Problem 10 Altıgen Çizimi

Kullanıcıdan altıgenin kaç satırdan oluşacağı ve orta satırında kaç karakter olacağı bilgisini alarak '*' karakteri ile bir altıgen çizin. (For loop kullanabilirsiniz.)

```python
satir_sayisi = int(input("Altıgen kaç satırdan oluşacak? "))
orta_satir_karakter = int(input("Orta satırda kaç karakter olacak? "))

# Üst bölüm
for i in range(satir_sayisi // 2):
    # Boşlukları ekle
    for j in range(satir_sayisi // 2 - i):
        print(" ", end="")
    # Yıldızları ekle
    for j in range(i * 2 + 1):
        print("*", end="")
    print()

# Ortadaki satır
for i in range(orta_satir_karakter):
    print("*", end="")
print()

# Alt bölüm
for i in range(satir_sayisi // 2):
    # Boşlukları ekle
    for j in range(i + 1):
        print(" ", end="")
    # Yıldızları ekle
    for j in range((satir_sayisi // 2 - i) * 2 - 1):
        print("*", end="")
    print()
```
